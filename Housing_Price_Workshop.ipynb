```python
def clip_outliers_df(x):
    x = pd.DataFrame(x).copy()
    for col in x.columns:
        Q1 = x[col].quantile(0.25)
        Q3 = x[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        median = x[col].median()
        x.loc[(x[col] < lower) | (x[col] > upper), col] = median
    return x.values

outliers_transformer=FunctionTransformer(clip_outliers_df, feature_names_out="one-to-one")
```

```python
from sklearn.pipeline import Pipeline

pip_num=Pipeline([
    ('num_imputer', SimpleImputer(strategy='median')), 
    ('Outlier_Handling', outliers_transformer ),
    ('RobustScaler', RobustScaler())
])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.ensemble import ExtraTreesRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRFRegressor

models = {
    "Linear": LinearRegression(),
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "ElasticNet": ElasticNet(),
    'XGBRFRegressor' : XGBRFRegressor(),
    "DecisionTree": DecisionTreeRegressor(),
    "RandomForest": RandomForestRegressor(n_estimators=200),
    "GradientBoosting": GradientBoostingRegressor(),
    "ExtraTrees": ExtraTreesRegressor(),
    "AdaBoost": AdaBoostRegressor(),
    "SVR": SVR(),
    "KNN": KNeighborsRegressor()
}
```

```python
preprocessor=ColumnTransformer([('num', pip_num, num_columns ),
                                ('cat_bin', bin_cat_pipline , bin_cat),
                                ( 'cat_ordinal',remaining_cat_pipline , remaining_cat)])
```

```python
from sklearn.metrics import mean_squared_error

results={}
for name, model in models.items():
    pipline=Pipeline([('preprocessor', preprocessor),
                      ('model', model)])
    
    pipline.fit(x_train, y_train)
    y_pred=pipline.predict(x_val)
    mse=mean_squared_error(y_val, y_pred)
    results[name]=mse
    print(f'model name: {name}      MSE= {mse:.4f}')
```

```python
def clip_outliers_y(y):
    y = pd.Series(y.ravel())  # (1000,1) 2D  to (1000,) 1D

    Q1 = y.quantile(0.25)
    Q3 = y.quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    median = y.median()

    y[(y < lower) | (y > upper)] = median

    return y.values.reshape(-1, 1)     #(1000,1) 2D
```

```python
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.pipeline import Pipeline

target_pipeline = Pipeline([
    ('outliers', FunctionTransformer(clip_outliers_y)),
    ('scaler', StandardScaler())
])
```

```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.metrics import mean_squared_error

results = {}
best_model=None
best_name=None
best_score=float('inf')

for name, model in models.items():
    
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    final_model = TransformedTargetRegressor(
        regressor=pipeline,
        transformer=target_pipeline,
        inverse_func=None
    )
    
    final_model.fit(x_train, y_train)
    
    y_pred = final_model.predict(x_val)
    
    mse = mean_squared_error(y_val, y_pred)
    results[name] = mse
    
    print(f"model name: {name}   MSE = {mse:.4f}")

    if mse < best_score:
        best_score=mse
        best_name=name
        best_model=final_model
```
